```python
# funcion para graficar la distribucion de los datos

def plot_distribution(data, cols=5, width=20, height=15, hspace=0.4, wspace=0.5):
    """
    Plots distribution of first few columns of a pandas dataframe.
    cols: number of columns to plot
    width: total width of the plot
    height: total height of the plot
    hspace: horizontal space between plots
    wspace: vertical space between plots
    """
    # compute rows required
    rows = math.ceil(float(data.shape[1]) / cols)
    # create a new figure with a certain size
    plt.figure(figsize=(width, height))
    # set the spacing between axes
    plt.subplots_adjust(top=0.99, bottom=0.01, hspace=hspace, wspace=wspace)
    # For each column, select the column and the plot
    for i, column in enumerate(data.columns):
        # compute the position of the subplot
        plt.subplot(rows, cols, i + 1)
        # plot the column's distribution
        sns.distplot(data[column],hist=False)
        # set the title of the plot
        plt.title(column)

```

```python
def filtrado_grande(df,pais,year_min,year_max):
    df = df[['Disaster Subgroup','Year','Country','Total Deaths','Total Affected','No Injured','No Homeless','No Affected']]
    df_filtrado = df[(df['Country'] == pais) & (df['Year'] >= year_min) & (df['Year'] <= year_max)][['Disaster Subgroup','Total Deaths','Total Affected','No Injured','No Homeless','No Affected']].groupby('Disaster Subgroup').sum()
    return df_filtrado
```

```python
def filtrado_chico(df,subgrupo,pais,year_min,year_max):
    df = df[['Disaster Subgroup','Year','Country','Total Deaths','Total Affected','No Injured','No Homeless','No Affected']]
    df_filtrado = df[(df['Disaster Subgroup'].isin(subgrupo)) & (df['Country'] == pais) & (df['Year'] >= year_min) & (df['Year'] <= year_max)][['Year','Total Deaths','Total Affected','No Injured','No Homeless','No Affected']].groupby('Year').sum()
    return df_filtrado
```
